```python
import sqlite3
import cv2
import face_recognition
import numpy as np
import matplotlib.pyplot as plt
from keras.models import load_model
model=load_model('face_recognition_facenet.h5')
img='muhammed_ali.jpg'
show_img=cv2.imread(img)
show_img=cv2.cvtColor(show_img,cv2.COLOR_BGR2RGB)
plt.imshow(show_img)
img1=face_recognition.load_image_file(img)
face_enc=face_recognition.face_encodings(img1)[0]
face_enc=face_enc.reshape(1,128)
preds=model.predict(face_enc)
print(preds)
result=np.argmax(preds,axis=1)[0]
print(result)
conn = sqlite3.connect("face_db.db")
cur=conn.cursor()
cursor=cur.execute("SELECT f_name FROM faces where f_id="+str(result+2))
for row in cursor:
    name=row[0]
name=name.replace('_',' ')
print("Hi, "+name)
```

```python
images_dir='lfw-deepfunneled'
import os
img_dir=os.listdir(images_dir)
import cv2
face_encodings_test=[]
i=-1
count=0
cls_labels=[]
for img in img_dir:
    i+=1
    for images in os.listdir(images_dir+'/'+img):
        try:
            img1=face_recognition.load_image_file(images_dir+'/'+img+'/'+images)
            face_encoding=face_recognition.face_encodings(img1)[0]
            face_encodings_test.append(face_encoding)
            cls_labels.append(i)
            count+=1
            print(count)
        except IndexError as e:
            pass
```

```python
face_encodings_test=np.asarray(face_encodings_test)
face_encodings_test=face_encodings_test.reshape((face_encodings_test.shape[0],128))
print(face_encodings_test.shape)
```

```python
predictions=model.predict(face_encodings_test)
```

```python
from sklearn.metrics import confusion_matrix,classification_report
print(classification_report(np.argmax(predictions,axis=1),np.array(cls_labels)))
print(confusion_matrix(np.argmax(predictions,axis=1),np.array(cls_labels)))
```
